```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your CSV data
df = pd.read_csv("alumni2023.csv")

counts = df.groupby(["BRANCH           /DIV", "Consent"]).size()

pivot_table = counts.unstack(level=-1)
```

```python
# Fill missing
pivot_table = pivot_table.fillna(0)
```

```python
# Plot the results
pivot_table.plot(kind="bar", stacked=True)
plt.title("Consent Distribution by Branch")
plt.xlabel("Sr. No. /DIV")
plt.ylabel("Count")
plt.legend(title="Consent")
plt.show()
```

```python
student_counts = df["BRANCH           /DIV"].value_counts()
student_counts.plot(kind="bar")
plt.title("Student Distribution by Division")
plt.xlabel("BRANCH/DIV")
plt.ylabel("Count")
plt.show()
```

```python
# Consent Trends
consent_counts = df["Consent"].value_counts()
consent_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Consent Trends")
plt.show()
```

```python
# Employer Preferences
employer_counts = df["Name of the Employer"].value_counts()
employer_counts.plot(kind="bar")
plt.title("Employer Preferences")
plt.xlabel("Name of the Employer")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
employer_counts = df["Name of the Employer"].value_counts()
```

```python
# Get the top 10 employers
top_10_employers = employer_counts.head(10)
```

```python
# Create a bar chart
top_10_employers.plot(kind="bar")
plt.title("Top 10 Employers for year 2023")
plt.xlabel("Name of the Employer")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# file_path = '/content/filetered2023data.csv'
df = pd.read_csv("filetered2023data.csv")
print(df.head())
print(df.info())
print(df.describe(include='all'))
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(f"Missing values in each column:\n{missing_values}")
```

```python
# Unique counts in columns
branches = df['Branch/ Div          '].nunique()
students = df[' Student Name '].nunique()
Consent_statuses = df['Consent '].value_counts()
interest_areas = df['INTEREST AREA'].value_counts()
categories = df['Category'].value_counts()
```

```python
# Display basic insights
print(f"Number of unique branches: {branches}")
print(f"Number of unique students: {students}")
print(f"Consent  distribution:\n{Consent_statuses}")
print(f"Top 5 interest areas:\n{interest_areas.head()}")
print(f"Category distribution:\n{categories}")
```

```python
# Visualize Data Insights
# 1. Distribution of students across branches
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Branch/ Div          ', order=df['Branch/ Div          '].value_counts().index)
plt.title('Distribution of Students Across Branches')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Consent  status visualization

sns.countplot(data=df, x='Consent ', order=df['Consent '].value_counts().index)
plt.title('Consent  Status Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Most common interest areas
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='INTEREST AREA', order=df['INTEREST AREA'].value_counts().index)
plt.title('Interest Areas Distribution')
plt.show()
```

```python
# 4. Category-wise student distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Category', order=df['Category'].value_counts().index)
plt.title('Category-Wise Student Distribution')
plt.xticks(rotation=45)
plt.show()
```
